```python
import import_ipynb
import matplotlib.pyplot as plt
import k_means, affinity_propagation, dbscan, divisive, optics, spectral
```

Visualization

Before Clustering

```python
embeddings_2d = k_means.embeddings_2d
vehicle_indices = k_means.vehicle_indices
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])
plt.title("Before Clustering")
plt.show()
```

K-Means Clustering

```python
vehicle_cluster_labels = k_means.vehicle_cluster_labels
tower_cluster_labels = k_means.tower_cluster_labels
plt.scatter(embeddings_2d[:len(vehicle_indices), 0], embeddings_2d[:len(vehicle_indices), 1], c=vehicle_cluster_labels, cmap='viridis', marker='o', label='Vehicles')
plt.scatter(embeddings_2d[len(vehicle_indices):, 0], embeddings_2d[len(vehicle_indices):, 1], c=tower_cluster_labels, cmap='viridis', marker='^', label='Towers')
plt.title("K-Means Clustering")
plt.show()
```

Accuracy Score Analysis

```python

k_means_score = k_means.accuracy
affinity_propagation_score = affinity_propagation.accuracy
dbscan_score = dbscan.accuracy
divisive_score = divisive.accuracy
optics_score = optics.accuracy
spectral_score = spectral.accuracy

accuracy_scores = [k_means_score, affinity_propagation_score, dbscan_score, divisive_score, optics_score, spectral_score]
print(accuracy_scores)
```
